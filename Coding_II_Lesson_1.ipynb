## Or much fancier, sciency things:

```
from matplotlib.collections import PolyCollection
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
import numpy as np

# Fixing random state for reproducibility



def polygon_under_graph(xlist, ylist):
    """
    Construct the vertex list which defines the polygon filling the space under
    the (xlist, ylist) line graph.  Assumes the xs are in ascending order.
    """
    return [(xlist[0], 0.), *zip(xlist, ylist), (xlist[-1], 0.)]


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make verts a list such that verts[i] is a list of (x, y) pairs defining
# polygon i.
verts = []

# Set up the x sequence
xs = np.linspace(0., 10., 26)

# The ith polygon will appear on the plane y = zs[i]
zs = range(4)

for i in zs:
    ys = np.random.rand(len(xs))
    verts.append(polygon_under_graph(xs, ys))

poly = PolyCollection(verts, facecolors=['r', 'g', 'b', 'y'], alpha=.6)
ax.add_collection3d(poly, zs=zs, zdir='y')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim(0, 10)
ax.set_ylim(-1, 4)
ax.set_zlim(0, 1)
plt.title("Are you impressed, {}?".format(name))

plt.show()
```

## What are RegExp?
- As the comic above indicates, its a syntax for searching specific patterns in text (an address, URL, emails, numbers) <br>
- It greatly simplifies the code that we could write using standard python string methods like split(), startswith(), endswith(), isalpha(), isdigit(), find() 
- They can become quite complex, entire books have been written about this subject. We will only cover the basics. https://en.wikipedia.org/wiki/Regular_expression
- The power of the regular expressions comes when we add special characters to the search string that allow us to more precisely control which lines match the string. Adding these special characters to our regular expression allow us to do sophisticated matching and extraction while writing very little code.
- They are not just Python specific, so other programming languages can have regex. Of course we will focus on the Python implementation


```
# here is the first code that deals with regex:
import re

text = "I want to search inside this text. This is just an example. Let's see if I can find the word: I"

found = re.search("I", text)
if found:
  print("I found the word 'I' on position ", found.start())
else: 
  print("I did not find 'I'")

found = re.search("We", text)
if found:
  print("I found the word 'We' on position ", found.start())
else: 
  print("I did not find 'We'")

```

So this is the basic functionality: Find the first occurrence of the token inside the string.<br>
The first steps were:

1.   Import re (re stands for regular expression and it is a standar library)
2.   Call the **search** function from the **re** library. This function takes 2 parameters: the regular expression to apply (the pattern to search) - in this first example it is just the letter 'I' and the second parameter is where to seach: in this example we have defined a simple text variable. Most of the times, you will want to search inside some external data source like a word file, pdf file, web page, etc.
3.   The function returns an object called a **Match Object**. If no match was found, the object is empty. We can use if to check if this object is empty or not. If it was found, this **match** object contains the information we care about like: at what position did the match occurred.


This is somewhat useful, but we can improve.
First improvement: let's find all the occurrences.

```
# here is the first code that deals with regex:
import re

text = "I want to search inside this text. This is just an example. Let's see if I can find the word: I"

found = re.finditer("I", text)
if found:
  print("I found the word 'I' {}  times".format(len(list(found))))
else: 
  print("I did not find 'I'")


```

In order to find all occurrences we call the function **finditer**.<br>
As the name suggests, it return an *iterator* object that has all the matches.<br>
Think of an iterator object as a list of **Matched Objects**. So as all the lists, we can take each matched object and print useful information.<br>
Let's combine the first 2 examples:

```
# here is the first code that deals with regex:
import re

text = "I want to search Inside this text. This is just an example. Let's see if I can find the word: I"

found = re.finditer("I", text)
if found:
  counter = 0
  for each_match in found:
    print("Occurence number {} of 'I' was on position {}".format(counter, each_match.start()))
    counter += 1

else: 
  print("I did not find 'I'")


```

Let's take it a step further: <br>
RegEx would not be so useful if we could only search for static tokens (a text that can only have one meaning). <br>
RegEx allows us to define patterns such as: 
- an email address
- a date
- a telephone number
- a valid card number

So they are used as both a way to search for information, but also to validate input. Think of a website that asks for your email address. How does it know if **bratiu@ie.edu** is valid, but **bratiu-ie.edu** is no valid? The answer is of course RegEx :) 

##Metacharacters and Forming Patterns
Some characters, like '|' or '(', are special. Special characters either stand for classes of ordinary characters, or affect how the regular expressions around them are interpreted.<br>
Here are the most common pattern conventions:
- **a, X, 9, <** -- ordinary characters just match themselves exactly. The meta-characters which do not match themselves because they have special meanings are: . ^ $ * + ? { [ ] \ | ( )

- **.** (a period) -- matches any single character except newline **'\n'**

- **\w** -- (lowercase w) matches a "word" character: a letter or digit or underbar [a-zA-Z0-9_]. 
- **\b** -- boundary between word and non-word
- **\t, \n, \r** -- tab, newline, return
- **^** = start, **$** = end -- match the start or end of the string
- **\** -- inhibit the "specialness" of a character. So, for example, use \\. to match a period or \\\ to match a slash. 

Another set of conventions are the repetitions:
- **\+** -- 1 or more occurrences of the pattern to its left, e.g. 'i+' = one or more i's
- **\*** -- 0 or more occurrences of the pattern to its left
- **?** -- match 0 or 1 occurrences of the pattern to its left
- **{x}**  match x occurrences of the pattern to its left
- **{x,y}**  match between x and y occurrences of the pattern to its left

```
# here are some examples:
## i+ = one or more i's, as many as possible.

#1
match = re.search(r'pi+', 'piiig') # found, match.group() == "piii"
print(1, match.group())

#2
match = re.search(r'i+', 'piigiiii') # found, match.group() == "ii", just the first one!
print(2, match.group())

#2-bis
match = list(re.finditer(r'i+', 'piigiiii')) # found, match.group() == "ii", just the first one!
print("2-bis", match[1].group()) # second match


#3
## ^ = matches the start of string, so this fails:
match = re.search(r'^bar', 'foobar') # not found, match == None
print(3, match)

#4
## but without the ^ it succeeds:
match = re.search(r'b\w+', 'foobar') # found, match.group() == "bar
print(4, match.group())

#5
match = re.search(r'(pig){2}', 'i have a pig and a pigpig') # found, match.group() == "pigpig"
print(5, match.group())

#6
match = re.search(r'\bpig\b', 'i have a pig and a pigpig') # found, match.group() == "pig"
print(6, match.group())

#7
match = re.search(r'.*', 'i have a pig and a pigpig') # this one matches all the string on a line
print(7, match.group())

```

## Grouping into sets
The last improvement we want to talk about are **[]**: <br>
- They are used to indicate a set of chars, so **[abc]** matches 'a' or 'b' or 'c'. <br>
- You can use **** to indicate a sequence of characters **[a-z]**  all the lowercase characters <br>
- Special characters lose their special meaning inside sets. For example, [(+*)] will match any of the literal characters '(', '+', '*', or ')'.

- Characters that are not within a range can be matched by complementing the set. If the first character of the set is '^', all the characters that are not in the set will be matched. For example, [^5] will match any character except '5', and [^^] will match any character except '^'. ^ has no special meaning if its not the first character in the set.

- To match a literal ']' inside a set, precede it with a backslash, or place it at the beginning of the set. For example, both [()[\]{}] and []()[{}] will both match a parenthesis.
- Repetition symbols can be used after a set.

**findall** is a simpler case of **finditer**. It just returns a list of all the matches. Sometimes it is all that we need.<br>


## Examples

```
text = "I want to search Inside this text. This is just an example. Let's see if I can find the word: I. My name is Bogdan Ratiu and I like to teach spectacular things"

#let's find all the 4 letter words:
words = re.findall(r"\b[\w]{4}\b", text)
print(words)

# what is we omit the \b ? 
words = re.findall(r"[\w]{4}", text)
print("without the \b we have: ", words)

# let us find all the proper names. A proper name is a collection of 2 words starting with a capital letter and with a space betweeen
names = re.findall(r"\b[A-Z]\w+ [A-Z]\w+\b", text)
print("Proper names:", names)

# match any number between 0-255
text = "122 345 1234 098 23 7 99 553"
expr = r"\b([0-9]|[1-9][0-9]|1[0-9][0-9]|2[0-4][0-9]|25[0-5])\b"
matched = re.findall(expr, text)
print("The numbers in the range 0-255 are", matched)

```

## Find the email addresses inside a word document

```
import re
import docx
from urllib.request import urlretrieve

#make sure that email contain @ and . after and there are no spaces in the middle
#classic way of getting the emails, without regex, quite complicated and not as good
def find_emails_classic(str):
    start = 0
    poz_start = 0
    emails_list = []
    while True:
        poz_at = str.find("@", start)
        if poz_at == -1:
            break

        poz_point = 0
        while poz_point < poz_at:
            poz_point = str.find(".", poz_point + 1)
            if poz_point == -1:
                break

        tmp = str.find(" ", start)
        while tmp < poz_at:
            poz_start = tmp + 1
            tmp = str.find(" ", poz_start)
            if tmp == -1:
                break

        #discard fake emails with spaces like: aaa@bbb ccc.dd
        poz_end = str.find(" ", poz_at)
        if poz_end < poz_point and poz_end != -1:
            start = poz_end
            continue

        poz_end = str.find(" ", poz_point)

        if poz_end == -1:
            emails_list.append(str[poz_start:])
        else:
            emails_list.append(str[poz_start: poz_end])

        start = poz_end
    return emails_list

#find emails via regex. Notice how much simpler it is:
reg_exp = "[a-zA-Z0-9]+@[a-zA-Z0-9]+\.[a-zA-Z0-9]+"
def find_emails_regex(str):
    return re.findall(reg_exp, str)

#make sure you have the right path to the DOCX file! We use urlretrieve
path = "https://github.com/itb-ie/MasterLaw-Session1-2/raw/master/Merger%20Agreement.docx"
filename2 = "emails.docx"
urlretrieve(path, "contract.docx")
doc = docx.Document("contract.docx")

all_emails = []
for para in doc.paragraphs:
    list_emails = find_emails_regex(para.text)
    for email in list_emails:
        all_emails.append(email)

#we have the emails
print("we now print the all_emails")
print(all_emails)
print("we are done")

doc = docx.Document()
para = doc.add_paragraph()
for em in all_emails:
    run = para.add_run(em)
    run.add_break()

doc.save(filename2)
```
